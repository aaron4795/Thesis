```python
# Import libraries
import pandas as pd
import os
import numpy as np
import math

# Set working directory
os.chdir('C:\\Users\\aaron\\Desktop\\GitHub\\Thesis\\Python')
```

```python
# Import data, just one year
time = ['9899', '9900', '0001', '0102', '0203', '0304', '0405', '0506', '0607', 
        '0708', '0809', '0910', '1011', '1112', '1213', '1314', '1415', '1516', 
        '1617', '1718', '1819', '1920', '2021', '2122']

dataframes = {}

for i in time:
    filename = f'SP1{i}.csv'
    
    filepath = os.path.join(r'Data\FootballUK', filename)
    
    # Import data and keep variables of interest
    df = pd.read_csv(filepath)
        
    if i == '9899' or i == '9900' or i == '0001' or i == '0102' or i == '0203' or i == '0304' or i == '0405':
        df = df[['Date', 'HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG']]
    else:
        df = df[['Date', 'HomeTeam', 'AwayTeam', 'FTR', 'FTHG', 'FTAG', 'HF', 'AF', 'HY', 'AY', 'HR', 'AR']]
    
    df['Season'] = i

    dataframes[filename] = df
```

```python
data = pd.concat(dataframes.values(), ignore_index = True)
```

```python
# Generate Year variable and fill according to the Season 
data.loc[data['Season'] == '9899', 'Year'] = 1998
data.loc[data['Season'] == '9900', 'Year'] = 1999
data.loc[data['Season'] == '0001', 'Year'] = 2000
data.loc[data['Season'] == '0102', 'Year'] = 2001
data.loc[data['Season'] == '0203', 'Year'] = 2002
data.loc[data['Season'] == '0304', 'Year'] = 2003
data.loc[data['Season'] == '0405', 'Year'] = 2004
data.loc[data['Season'] == '0506', 'Year'] = 2005
data.loc[data['Season'] == '0607', 'Year'] = 2006
data.loc[data['Season'] == '0708', 'Year'] = 2007
data.loc[data['Season'] == '0809', 'Year'] = 2008
data.loc[data['Season'] == '0910', 'Year'] = 2009
data.loc[data['Season'] == '1011', 'Year'] = 2010
data.loc[data['Season'] == '1112', 'Year'] = 2011
data.loc[data['Season'] == '1213', 'Year'] = 2012
data.loc[data['Season'] == '1314', 'Year'] = 2013
data.loc[data['Season'] == '1415', 'Year'] = 2014
data.loc[data['Season'] == '1516', 'Year'] = 2015
data.loc[data['Season'] == '1617', 'Year'] = 2016
data.loc[data['Season'] == '1718', 'Year'] = 2017
data.loc[data['Season'] == '1819', 'Year'] = 2018
data.loc[data['Season'] == '1920', 'Year'] = 2019
data.loc[data['Season'] == '2021', 'Year'] = 2020
data.loc[data['Season'] == '2122', 'Year'] = 2021

# Standarize the names of the teams
team = ['HomeTeam', "AwayTeam"]
for i in team:
    data[i] = data[i].str.replace('Alaves', 'Alavés')
    data[i] = data[i].str.replace('Almeria', 'Almería')
    data[i] = data[i].str.replace('Ath Bilbao', 'Athletic Club')
    data[i] = data[i].str.replace('Ath Madrid', 'Atlético Madrid')
    data[i] = data[i].str.replace('Cadiz', 'Cádiz')
    data[i] = data[i].str.replace('Celta', 'Celta Vigo')
    data[i] = data[i].str.replace('Córdoba CF', 'Córdoba')
    data[i] = data[i].str.replace('Espanol', 'Espanyol')
    data[i] = data[i].str.replace('Gimnastic', 'Gimnàstic')
    data[i] = data[i].str.replace('Hercules', 'Hércules')
    data[i] = data[i].str.replace('La Coruna', 'La Coruña')
    data[i] = data[i].str.replace('Leganes', 'Leganés')
    data[i] = data[i].str.replace('Malaga', 'Málaga')
    data[i] = data[i].str.replace('Murcia', 'Real Murcia')
    data[i] = data[i].str.replace('Real Oviedo', 'Oviedo')
    data[i] = data[i].str.replace('Santander', 'Racing Sant')
    data[i] = data[i].str.replace('Sociedad', 'Real Sociedad')
    data[i] = data[i].str.replace('Sp Gijon', 'Sporting Gijón')
    data[i] = data[i].str.replace('Vallecano', 'Rayo Vallecano')
```

```python
data.rename(columns={'FTHG': 'FinalGHome', 'FTAG' : 'FinalGAway', 
                     'FTR' : 'FullTime', 'HF' : 'homefouls',
                     'AF' : 'awayfouls', 'AY' : 'awayYellow',
                     'HY' : 'homeYellow', 'AR' : 'awayRed',
                     'HR' : 'homeRed'}, inplace= True)
```
